```python
#exeption handling
count=0    #ounter
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[count, 'OWN_OCCUPIED']=np.nan
        
    except ValueError:
            pass
    count= count + 1
```
